```python
from scipy import optimize

# 최소화할 함수 정의
def func(x):
    return (x - 3) ** 2

# 함수의 최소값 찾기
result = optimize.minimize(func, x0=0)
print("Minimum value:", result.fun)  # 출력: Minimum value: 0.0
```

```python
from scipy import optimize

def func(x):
    return (x - 3) ** 2

# 다른 초기 추정값들
initial_guesses = [0, 10, -5]

for guess in initial_guesses:
    result = optimize.minimize(func, x0=guess)
    print(f"Initial guess: {guess}, Minimum value: {result.fun}, x at minimum: {result.x[0]}")
```

```python
import numpy as np
from scipy import optimize

def func(x):
    return np.sin(x) ** 2

# 다중 최솟값을 가진 함수
initial_guesses = [0, 2, 5]

for guess in initial_guesses:
    result = optimize.minimize(func, x0=guess)
    print(f"Initial guess: {guess}, Minimum value: {result.fun}, x at minimum: {result.x[0]}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize

def func(x):
    return np.sin(x) ** 2

# 다중 최솟값을 가진 함수
x = np.linspace(-2 * np.pi, 2 * np.pi, 400)
y = func(x)

# 초기 추정값
initial_guesses = [0, 2, 5]

# 결과 저장
results = []
for guess in initial_guesses:
    result = optimize.minimize(func, x0=guess)
    results.append((guess, result.fun, result.x[0]))

# 시각화
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='func(x) = sin(x)^2', color='blue')
plt.scatter([res[2] for res in results], [res[1] for res in results], color='red', zorder=5, label='Minima found')
plt.axhline(y=0, color='black', linestyle='--', linewidth=0.7)
plt.title('Visualization of func(x) = sin(x)^2 with Optimization Results')
plt.xlabel('x')
plt.ylabel('func(x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import cv2
import pytesseract
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# Tesseract 경로 설정 (Colab에서는 경로 설정이 필요 없음)
# pytesseract.pytesseract.tesseract_cmd = r'/usr/bin/tesseract'

def preprocess_image(image_path):
    # 이미지 로드
    image = cv2.imread(image_path)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    blurred = cv2.GaussianBlur(gray, (5, 5), 0)
    edged = cv2.Canny(blurred, 30, 150)
    return edged

def detect_license_plate(image):
    # 번호판 영역 찾기
    contours, _ = cv2.findContours(image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    for contour in contours:
        x, y, w, h = cv2.boundingRect(contour)
        if w > 100 and h > 30:  # 번호판 크기 조건 (임의로 설정)
            return x, y, w, h
    return None

def extract_text_from_plate(image, plate_rect):
    x, y, w, h = plate_rect
    plate_image = image[y:y+h, x:x+w]
    text = pytesseract.image_to_string(plate_image, config='--psm 8')
    return text.strip()

# 이미지 업로드
uploaded = files.upload()

# 첫 번째 업로드된 파일 경로
image_path = list(uploaded.keys())[0]

# 이미지 전처리
preprocessed_image = preprocess_image(image_path)

# 원본 이미지 로드
original_image = cv2.imread(image_path)

# 번호판 탐지
plate_rect = detect_license_plate(preprocessed_image)
if plate_rect:
    x, y, w, h = plate_rect
    cv2.rectangle(original_image, (x, y), (x+w, y+h), (0, 255, 0), 2)

    # 번호판 문자 인식
    text = extract_text_from_plate(original_image, plate_rect)
    print("Detected License Plate Text:", text)
else:
    print("License plate not detected.")

# 결과 시각화
plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
plt.title('License Plate Detection')
plt.axis('off')
plt.show()
```
